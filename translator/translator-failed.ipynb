```python
from azure.core.credentials import AzureKeyCredential
from azure.ai.translation.text import TextTranslationClient
```

```python
apikey = ""
region = "westus"
endpoint = "https://api.cognitive.microsofttranslator.com/"
```

```python
credential = AzureKeyCredential(apikey)
translator_client = TextTranslationClient(credential=credential,endpoint=endpoint)
```

```python
def translate_text(text, to_language="es"):
    response = translator_client.translate(content=[text], to=[to_language])
    return response[0].translations[0].text

translated_text = translate_text("Hello world", "ja")
print(translated_text)
```
