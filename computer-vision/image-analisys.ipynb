```python
import os
from azure.ai.vision.imageanalysis import ImageAnalysisClient
from azure.ai.vision.imageanalysis.models import VisualFeatures
from azure.core.credentials import AzureKeyCredential
from dotenv import load_dotenv
```

```python
load_dotenv()
endpoint = os.environ["endpoint"]
key = os.environ["key"]
```

```python
client = ImageAnalysisClient(
        endpoint=endpoint,
        credential=AzureKeyCredential(key)
    )
```

```python
with open("splinter-cell-promo-image.jpg", "rb") as f:
    image_bytes = f.read()
```

```python
result = client.analyze(
        image_data=image_bytes,
        visual_features=[VisualFeatures.CAPTION, VisualFeatures.DENSE_CAPTIONS],
        gender_neutral_caption=False  # Set to True if required by compliance
    )
```

```python
if result.caption is not None:
        print("--- Primary Caption ---")
        print(f"Description: '{result.caption.text}'")
        print(f"Confidence: {result.caption.confidence:.4f}\n")
```

```python
if result.dense_captions is not None:
        print("--- Regional Detailed Descriptions ---")
        for idx, caption in enumerate(result.dense_captions.list):
            box = caption.bounding_box
            print(f"Region {idx + 1}: '{caption.text}'")
            print(f"  Confidence: {caption.confidence:.4f}")
            print(f"  Bounding Box: [X: {box.x}, Y: {box.y}, Width: {box.width}, Height: {box.height}]")
```
